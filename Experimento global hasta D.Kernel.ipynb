```python
from geopy.geocoders import Nominatim as loc
import pandas as pd
import boto3
```

```python
geo = loc(user_agent = 'enzusa09@gmail.com')
comprehend=boto3.client(service_name='comprehend', region_name='us-west-2')
```

```python
df = pd.read_csv("EXPERIMENTO.csv",encoding="utf-8") 
```

```python
A = df['Embedded_text']
```

```python
AT121 = []
```

Amazon Comprehend

```python
for e in A:
    try:
        json1 = comprehend.detect_entities(Text=e, LanguageCode='es')
        for entidad in json1['Entities']:
            if entidad['Type'] == 'LOCATION':
                AT121.append([e,entidad['Text']])
    except:
        pass
```

```python
AsaltoT121 = pd.DataFrame(AT121,columns=["Tweet","Locacion"])
AsaltoT121.to_csv("Entidades Experimento.csv",encoding="UTF-8")
```

```python
df = pd.read_csv("Entidades Experimento.csv",encoding="utf-8") 
```

```python
A = df['Locacion']
```

Direcciones

```python
coordenadasA = []
z=1
for e in A:
    try:
        location = geo.geocode(e)
        direccion = location.address
        if 'Ciudad de México' in direccion:
            coordenadasA.append([direccion,z])
    except:
        pass
```

```python
AsaltoT121 = pd.DataFrame(coordenadasA,columns=["Direccion","Cantidad"])
Datos = AsaltoT121.groupby("Direccion").sum()
Datos.to_csv("Conjunto de Direcciones Experimento.csv",encoding="UTF-8")
```

```python
df = pd.read_csv("Conjunto de Direcciones Experimento.csv",encoding="utf-8") 
```

```python
A = df['Direccion']
SA = df['Cantidad']
```

Coordenadas

```python
coordenadasA = []
for a, b in zip(A,SA):
    location = geo.geocode(a)
    coordenadasA.append([a,location.longitude,location.latitude,b])
```

```python
AsaltoT121 = pd.DataFrame(coordenadasA,columns=["Direccion","Longitud","Latitud","Cantidad"])
AsaltoT121.to_csv("Coordenadas de Experimento.csv",encoding="UTF-8")
```
